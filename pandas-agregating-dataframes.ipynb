# Summary Statistics

numbers that sumarice and tell you about the dataset

`data_frame_name['columna_interes].` segido de uno de los siguientes metodos

`.mean()`: promedio

`.median()`: mediana

`.mode()`: moda

`.min()`

`.max()`

`.var()`: varianza

`.std()`: variacion estandar

`.sum()`: suma de los valores de esa columna

`.quantile()`: calcular quantiles

## fechas

si una columna contiene fechas en el formato 'año-mes-dia' asi AAAA-MM-DD, datatype `datetime64`

tambien se le puede aplicar funciones estadisticas como

`.min()`

`.max()`

## .agg() method

sirve para calcular custom summary estatistics

```python
import pandas as pd

frame_soccer = pd.read_csv('datasets/soccer.csv', index_col=0)

def pct30(column:pd.Series):
    return column.quantile(0.3)

frame_soccer['height'].agg(pct30)
```

### .agg() para multiples columnas

```python

frame_soccer[['height', 'rating']].agg(pct30)
```

### .agg() para multiples funciones summary

se pasa una LISTA de funciones como parametro a `.agg(  )`

```python
def pct40(column:pd.Series):
    return column.quantile(0.4)

frame_soccer['height'].agg([pct30, pct40])
```

### multiples funcionesy multiples columnas

```python
import numpy as np
frame_soccer[['height', 'rating']].agg([pct30, pct40, np.mean]) 
```

### Cumulatives statistics methods

genera un acolumna calculando dichos valor con solo las filas recorridas hasta esa fila

```python
frame_soccer['rating'].cumsum()
```

`columna.cumsum()` : suma acumulada

`columna.cummin()` : minimo entre las columnas anteriores incluyendo la actual

`columna.cummax()` : maximo entre las columnas anteriores incluyendo la actual

`columna.cumprod()`:producto acumulado

# Counting

## Drop duplicates 

remover filas que contienen entradas repetidas en una columna

```python
frame_soccer.drop_duplicates # por si solo remueve filas repetidas, con todos sus atrubutos iguales

heigths = frame_soccer.drop_duplicates(subset='height')['height']
print(heigths.sort_values(ascending=True))
```

tambien se pueden considerar multiples columnas para el metodo `drop_duplicates` si dos o más filas tienen los mismos valores para una sub lista de columnas pasadas como parametro a este metodo, solo una fila sera conservada

```python
frame_soccer.drop_duplicates(subset=['rating', 'position'])
```

# Count

para contar el numero de entradas que tienen un valor de una columna especifico se extrae la columna, como un Pandas Series, y se llama al metodo `.value_counts()`. a este metodo tambien se le puede decir que ordene `(sort_True)`

contemos cuantos jugadores hay por cada posicion en la base de datos

```python
print(frame_soccer['position'].value_counts(sort=True))
```

# Proportions

al usar el metodo `.value_counts()` se puede hacer el conteo por proporciones del total en lugar de numero de registrosm, esto se hace al pasar como parametro `(normalize=True)`

```python
positions_proportion = frame_soccer['position'].value_counts(sort=True, ascending=False,  normalize=True)
print(positions_proportion, '\n')# Pandas Series

positions_data = pd.DataFrame(positions_proportion)
positions_data['percentage'] = positions_data['position']*100
print(positions_data) # Pandas DataFrame
```

# Gruopued Summary Statistics


tratemos de calcular la estatura promedio por cada Posicion de jugador

primero veamos que tipos de posiciones diferentes hay:

```python
frame_soccer.drop_duplicates(subset='position')['position']
```

Ahora calculamos la esatura promedio entre los jugadores de cada posicion

```python
print(frame_soccer[frame_soccer['position'] == 'GK']['height'].mean())
print(frame_soccer[frame_soccer['position'] == 'M']['height'].mean())
print(frame_soccer[frame_soccer['position'] == 'A']['height'].mean())
print(frame_soccer[frame_soccer['position'] == 'D']['height'].mean())
```

## GroupBy 
la anterior tarea se puede simplificar usando el metodo groupby de un dataframe. esto reduce las lineas de codigo y tambien da un output con nombres de valores agrupados y asi sabemos que valor corresponde a que

```python
frame_soccer.groupby('position')['height'].mean()
```

## Multiple Gruped Summaries

una vez agrupadas las estradas por un valor de una columna especifica se pueden calcular multiples atributos estadisticos usando `.agg([lista de funciones])`

```python
frame_soccer.groupby('position')['height'].agg([min, max, sum, pd.DataFrame.mean])
```

## Agrupar por multiples columnas

```python
# esto tambien es valido pero el output no sale en formato de tabla
#frame_soccer.groupby(['position', 'foot'])['height'].mean()
frame_soccer.groupby(['position', 'foot'])['height'].agg([pd.DataFrame.mean])
```

## Agregate por multiples columnas

```python
frame_soccer.groupby(['position', 'foot'])['height', 'rating'].agg([pd.DataFrame.mean])
```

## Agrupar por multiples columnas y Agregate por multiples funciones

```python
frame_soccer.groupby(['position', 'foot'])['height'].agg([max, min])
```

# Pivot Tables

son otra forma de calcular grouped summary statistics, por default este calcula el promedio de los valores 

```python
# frame_soccer.groupby('position')['height'].mean()
# tambien se puede lograr con:
frame_soccer.pivot_table(values='height', index='position')
```

si se requiere otra funcion aparte de promedio se asigna un valor al parametro `aggfunc`

```python
frame_soccer.pivot_table(values='height', index='position', aggfunc=np.median)
```

usando pivot tables se puede Agrupar por multiples columnas, hacer agregados sobre multiples columnas (columnas de calculo) con multiples funciones 

```python
frame_soccer.pivot_table(values=['height', 'rating'], index='position', aggfunc=[np.mean, np.min, np.max])
```

## Pivot en dos variables

 The first argument is the column name containing values to aggregate. The index argument lists the columns to group by and display in rows, and the columns argument lists the columns to group by and display in columns

para agrupacion con multiples columnas

```python
frame_soccer.pivot_table(values='height', index='position', columns='foot', aggfunc=np.mean)
```

# Filling Missing Values
donde haya un valo numerico, se puede agregar un valor por defecto

fill_value replaces missing values with a real value (known as imputation)

```python
frame_soccer.pivot_table(values='shooting', index='position', columns='rating', fill_value=0)
```

# Sumando con Pivot Tables

si se agregaa el atributo margins = True se agrega una fila y una columna con la el calculo de las entradas en esa fila/columna

```python
frame_soccer.pivot_table(values='height', index='position', columns='foot', aggfunc=np.mean, margins=True)
```

# Working with pivot tables: Calculations 

volvamos a trabajar con el dataframe tipo tabla pivot agrupando en las columnas los tipos de piernas diestras, en las filas agrupando por posicion, y el calculo de cada grupo es el promedio de altura

```python
mean_height_grouped_by_position_foot = frame_soccer.pivot_table(values='height', index='position', columns='foot', aggfunc=np.mean)
print(mean_height_grouped_by_position_foot)
```

se puede hacer slicing de filas y columnas con .loc y .iloc pero tambien se pueden hacer calculos sobre las filas y las columnas de una pivot table usando para ello el atributo `axis` de las funciones de calculos estadisticos

 The default value is "index," which means "calculate the statistic across rows |||"

```python
mean_height_grouped_by_position_foot.mean(axis='index')
```

To calculate a summary statistic for each row, that is, "across the columns _==" you set axis to "columns.

```python
mean_height_grouped_by_position_foot.mean(axis='columns')
```
