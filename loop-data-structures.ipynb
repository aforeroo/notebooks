# Loop over data structures

# use enumerate


```python
lista = ['uno', 'dos', 'tres', 'cuatro']
for index, name in enumerate(lista):
    print(index, '=', name)
```

se puede cambiar el inicio del indexador, por defecto es 0

```python
lista = ['uno', 'dos', 'tres', 'cuatro']
for index, name in enumerate(lista, start=1):
    print(index, '=', name)
```

# listas de listas

```python
# house list of lists
house = [["hallway", 11.25], 
         ["kitchen", 18.0], 
         ["living room", 20.0], 
         ["bedroom", 10.75], 
         ["bathroom", 9.50]]
         
# Build a for loop from scratch
for x, sqm in house:
    print('the',x,'is',sqm,'sqm')
```

# Diccionarios

se usa la funcion `dicccionario.items()` para obtener parajeas (llave, valor) asignadas a los nombres de los iteradores

```python
world = { "afghanistan":30.55, 
          "albania":2.77,
          "algeria":39.21 }


for key, value in world.items() :
    print(key + " -- " + str(value))
```

# Numpy Array

If you're dealing with a 1D NumPy array, looping over all elements can be as simple as:

`for x in my_array :`
 
    ...



If you're dealing with a 2D NumPy array, it's more complicated. A 2D array is built up of multiple 1D arrays. To explicitly iterate over all separate elements of a multi-dimensional array, you'll need this syntax:


`for x in np.nditer(my_array) :`

# Pandas DataFrame

```python
import pandas as pd
brics = pd.read_csv('datasets/brics.csv', index_col=0)
```

iterar sobre nombre de columnas

```python
for col in brics:
    print(col)
```

### Iterar valores de columnas

`values` es de tipo pandas.Series

```python
div_line = '\n' + '-'*20
for column_name, values in brics.iteritems():
    print(column_name,':')
    print(values)
    print(div_line)
```

### Iterar valores de filas

`atributes` es de tipo pandas.Series

```python
for row_id, atributes in brics.iterrows():
    print(row_id,':')
    print(atributes)
    print(div_line)
```

```python
for label, row in brics.iterrows():
    print(label, ':', row['capital'])
```

### iterar por Tuplas pandas

```python
for a in brics.itertuples():
    print(a.capital, 'is the capital of', a.country)
```
