# Data Merging basics

## Inner Join

dadas dos tablas que contengan una llave comun estas se pueden unir en un solo data frame del siguiente modo

`new_frame = left_frame.merge(rigth_frame, on='columna_llave')`


si ambas tablas tienen columnas con nombres iguales se busca que la abla final tenga sufios para poder identificarlas

el nombre de los sufijos se puede especificar con el parametro `suffixes=('_suffix_left','_siffix_rigth')`

`new_frame = left_frame.merge(rigth_frame, on='column_key', suffixes=('_suffix_left','_siffix_rigth'))`

## Left Join

a diferencia del inner join, un left join conserva todas las entradas de la tabla o data frame de la izquierda y agrega las columnas que tengan coincidencia con la llave a hacer merge, si un valor en la tabla derecha, no tiene un valor asociado a un id, en la resultado este campo aparecera como NAN o null o vacio simplemente

para hacer un left join con pandas dataframes se hace de la siguiente manera

`new_frame  = left_table.merge(rigth_table, on='id_key', how='left')` 

A left join will return all of the rows from the left table. If those rows in the left table match multiple rows in the right table, then all of those rows will be returned. Therefore, the returned rows must be equal to if not greater than the left table.
