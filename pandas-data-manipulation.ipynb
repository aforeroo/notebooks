# 1. Intro to DataFrames

Pandas se vale del uso de numpy y matplotlib, Los Dataframes son el core de pandas junto con Series

Un DataFrame se usa para almacenar informacion rectangular: Filas y columnas. Como lo haria una tabla de una base de datos relacional. Cada Columna puede contener un tipo de dato diferente

- para ver las columnas y las primeras filas del DataFrame se usa el metodo `.head(#_first_rows)`

- para ver la informacion de las columnas, sus nombres y numero de missing values se usa `.info()`

- si se quiere saber el numero de filas y de columnas se llama el atributo `.shape` que retorna una tupla en la forma `(#rows, #columns)`

- Para ver iformacion estadistica basica sobre las columnas numericas se usa `.describe()`


```python
import pandas as pd

frame_soccer = pd.read_csv('datasets/soccer.csv', index_col=0)
frame_soccer.head()
```

```python
print(frame_soccer.info())
```

```python
frame_soccer.shape
```

```python
frame_soccer.describe()
```

# Componentes de DataFrame

- The values attribute `.values`, as you might expect, contains the data values in a 2-dimensional NumPy array. SIn la fila de los nombres de columnas

- The columns attribute `.columns` contains column names into a index object, 
- the index attribute `.index` contains row numbers or row names as an index object

# Sorting & Subsetting

- ### Ordenar por una Columna

se usa el siguiente formato: `data_frame.sort_values('columna_de_ordenamiento', ascending=False)`

`ascending` por defecto esta en `True`, lo cual muestra de menor a mayor, para ver el ordenamiento de mayor a menor se usa `ascending=False`

```python
frame_soccer.sort_values('rating', ascending=False).head()
```

- ### Ordenar por multiples columnas

esta vez, en lugar de uno, al primer argumento del metodo `sort_values(...)` se le pasa una lista de nombres de columnas a ordenar por. 

```python
frame_soccer.sort_values(['height','rating'], ascending=[False, True]).head(10)
```

- ### Selecionando solo una Columna

despues del nombre, entre parentesis cuadrados simples ponenemos entre comillas el nombre de la columna a seleccionar. take a look at the type of this output

```python
frame_soccer['name']
```

```python
type(frame_soccer['name'])
```

se puede lograr lo mismo pero manteniendo el tipo de dato: DataFrame, al usar brackets dobles

```python
frame_soccer[['name']]
```

```python
type(frame_soccer[['name']])
```

## Seleccionar multiples columnas

```python
frame_soccer[['name', 'height']]
```

- ### Seleccionando Filas por una condición

```python
frame_soccer[ frame_soccer['height'] < 160 ]
```

- ### Seleccionando filas basados en texto

```python
frame_soccer[ frame_soccer['position'] == 'GK']
```

### Seleccionando filas por multiples condiciones

se usa `&` para el AND logico

```python
is_goal_keeper = frame_soccer['position'] == 'GK'
is_very_tall = frame_soccer['height'] > 199 
frame_soccer[is_goal_keeper & is_very_tall] 
```

## .isin() Method

en lugar de usar el OR logico `|` y si se quiere filtrar por multiples valores de una varibale categorica se usa isin() method con el siguiente formato

`sub_set = data_frame['categotical_column'].isin(['value1', ... ,'valueN'])`
_______

`colors = ["brown", "black", "tan"]`

`condition = dogs["color"].isin(colors)`

`dogs[condition]`

por ejemplo queremos todos los jugadores que sean arqueros `GK` OR mediocampistas `M`



```python
arqueros_medios = frame_soccer['position'].isin(['GK', 'M'])
frame_soccer[arqueros_medios]
```

# Nuevas columnas

se puede añadir una nueva columna que contenga datos calculados a partir de otras variables en la misma fila

por ejemplo queremos guardar la estatura de los jugadores de futbol en metros y tambien conservar la de estatura en centimetros

```python
frame_soccer['height_meters'] = frame_soccer['height']/100
frame_soccer[['name', 'height', 'height_meters']]
```
