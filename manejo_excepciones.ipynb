# Manejo de Excepciones

## Try, Except

```python
def fun(s:str):
    return s[::-1]
```

si pasamos como parametro algo distinto a un objeto str o currira una excepcion del tipo TypeError

```python
#n = input('ingrese una cadena de caracteres')
n = 2
try:
    print(fun(n))
except TypeError:
    print('Te dije que ingresaras una cadena de caracteres JOderrr')
```

## Raise
puede que nosotros consideremos que un caso donde consideramos que hay un error, pero que se sabe python no lo considera asi pues no hay un error en tiempo de ejecucion pero para la logica de nuestro programa si lo es

```python
#n = input('ingrese una cadena de caracteres no vacia')
n = ''
try:
    if len(n) == 0: # esto se puede hacer aca o en el codigo de la funcion
        raise ValueError('Por que no lees? te dije que no fuera vacia')
    print(fun(n))
except ValueError as ve:
    print(ve)
except TypeError:
    print('Te dije que ingresaras una cadena de caracteres JOderrr')
```

## Finally

se usa al final de una estructura Try, Except para cerrar un archivo, cerrar una conexion a una base de datos o liberar recursos externos


```python
try:
    f = open('package_requirement.txt') # dentro de un try porque abrir un archivo puede generar errores
finally:
    f.close()
```

## Assert Staments

`assert` condicion, 'mensaje de error'

```python
def fun(s:str):
    assert len(s) > 0, 'No se puede ingresar una cadena vacia'
    return s[::-1]

fun('')
```
