# Random Generators with Numpy

se usa el sub-paquete `random` del modulo `numpy`

```python
import numpy as np
print('un numero pseudo aleatorio entre 0 y 1 - semilla elegida por python al llamar a rand')
np.random.rand()
```

se puede elegir un numero entero como semilla con `np.random.seed(#int_semiila)` lo cual garantiza la Reproductibilidad, lo que significa que cualquiera que use la misma se milla puede replicar la serie de numeros aleatorios usados en un programa o simulaci√≥n (respetando el numero  y orden de llamdos a rand() o randint())

```python
np.random.seed(21)
print('primer numero en la serie:', np.random.rand())
print('segundo numero en la serie:', np.random.rand())

print('reset de la semilla al mismo valor')
np.random.seed(21)
print('primer numero en la serie:', np.random.rand())
print('segundo numero en la serie:', np.random.rand())

```

# Arreglos y Matrices de numeros enteros aleatorios

```python
# run this to see randint documentation
#help(np.random.randint)
```

para crear un arreglo de N numeros aleatorios enteros se usa el siguiente formato `np.random.randint(min, max+1, N)`

```python
min = 0
max  = 1
N = 5
print(f'arreglo de {N} numeros enteros entre {min} y {max}')

np.random.randint(min,max+1,N)
```

Para crear arreglos de N numeros enteros aleatorios en los que cada elementos tiene un min y maximo especifico se usa el siguiente formato:

    np.random.randint([min1, min2, ... , minN], [max1 + 1, max2 +1, ... , maxN + 1 ])

```python
print('primer valor entre 0 y 1')
print('segundo valor entre 1 y 2')
print('primer valor entre 2 y 3')

np.random.randint([0,1,2],[1 +1,2 +1,3 +1])
```

Para crear una matriz de N filas X M columnas en las que cada elemento esta entre min_global y max_gloval se usa el siguiente formato

`np.random.randint([[min_global]*M]*N,[max_global+1])`

```python
min_global = 0
max_global = 2
filas = 2
columnas = 3
print(f'matriz {filas} X {columnas} con valores entre {min_global} y {max_global}')

np.random.randint([[min_global]*columnas]*filas,[max_global+1])
```

Para crear una matriz de N filas X M columnas en las que cada elemento esta entre min_col y max_col se usa el siguiente formato

`np.random.randint( [[min_col0, min_col1, ... ,min_colM]]*N, [max_col0 +1, max.col1 + 1, ... , max_colM +1 ] )`

```python
# matriz 3x3 em la que los valores de la columna i estan entre i y i + 1 
print('min    ',0,'',1,'',2)
print('max    ',1,'',2,'',3)

np.random.randint( [[0,1,2]]*4, [1+1,2+1,3+1])
```

Para crear una matriz de N filas X M columnas en las que cada elemento esta entre min_row y max_row se usa el siguiente formato

`np.random.randint([[min_row0]*M,[min_row1]*M, ..., [min_rowN]*M], [[max_row0 + 1],[max_row1 + 1], ... , [max_rowN +1]])`

```python
np.random.randint([[0]*2,[1]*2,[2]*2], [[2],[3],[4]])
```

Para crear una matriz de N filas X M columnas en las que cada elemento esta entre min_ij y max_ij se usa el siguiente formato

`np.random.randint([[min_11, min12, ... , min1M], [min_21, min22, ... , min2M], ... , [minN1, minN2, ... minNM]],`
                   
`               [[max_11 +1, max12+1, ... , max1M+1], [max_21+1, max22+1, ... , max2M+1], ... , [maxN1+1, maxN2+1, ... maxNM+1]])`


```python
print(' 0 <= M[0][0] < 1' )
print(' 1 <= M[0][1] < 2' )
print(' 2 <= M[1][0] < 3' )
print(' 3 <= M[1][1] < 4' )
np.random.randint([[0,1],[2,3]], [[2,3],[4,5]])
```
