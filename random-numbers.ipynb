# Random Generators with Numpy

para crear un arreglo de N numeros aleatorios enteros se usa el siguiente formato `np.random.randint(min, max+1, N)`

```python
minimum = 0
maximum  = 1
N = 5
print(f'arreglo de {N} numeros enteros entre {minimum} y {maximum}')

np.random.randint(minimum,maximum+1,N)
```

Para crear arreglos de N numeros enteros aleatorios en los que cada elementos tiene un min y maximo especifico se usa el siguiente formato:

    np.random.randint([min1, min2, ... , minN], [max1 + 1, max2 +1, ... , maxN + 1 ])

```python
print('primer valor entre 0 y 1')
print('segundo valor entre 1 y 2')
print('primer valor entre 2 y 3')

np.random.randint([0,1,2],[1 +1,2 +1,3 +1])
```

Para crear una matriz de N filas X M columnas en las que cada elemento esta entre min_global y max_gloval se usa el siguiente formato

`np.random.randint([[min_global]*M]*N,[max_global+1])`

```python
min_global = 0
max_global = 2
filas = 2
columnas = 3
print(f'matriz {filas} X {columnas} con valores entre {min_global} y {max_global}')

np.random.randint([[min_global]*columnas]*filas,[max_global+1])
```

Para crear una matriz de N filas X M columnas en las que cada elemento esta entre min_col y max_col se usa el siguiente formato

`np.random.randint( [[min_col0, min_col1, ... ,min_colM]]*N, [max_col0 +1, max.col1 + 1, ... , max_colM +1 ] )`

# Random Walk

basado en una variable aleatoria, en este caso el lanzamiento de un dado, se avanza cierta cantidad de pasos(no muy grandes), ya sea arriba, abajo, izquierda, derecha dependiendo de dos cosas:
- la posicion anterior
- el resultado de la variable aleatoria

```python
# Initialize random_walk
random_walk = [0]

# Complete the ___
for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        step = step - 1
        #step = max(0, step - 1) si no se puede ir por debajo de cero
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)

#plot random_wak values vs their indexes
import matplotlib.pyplot as plt
plt.plot(random_walk)
plt.show()
```

Ahora el codigo completo

```python
final_steps = []
for i in range(10000):#entre mayor sea este numero mas se acercara a la distribucion teorica
    random_walk = [0]
    for i in range(100):
        # Set step: last element in random_walk
        step = random_walk[-1]

        # Roll the dice
        dice = np.random.randint(1,7)

        # Determine next step
        if dice <= 2:
            step = max(0,step-1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # append next_step to random_walk
        random_walk.append(step)
     
    final_steps.append(random_walk[-1])

plt.hist(final_steps, bins=15)
plt.show()    
    

```

## Graficar una lista de listas: ver y comparar diferentes random walks

ahora en lugar de guardar solo los ultimos valores, guardamos el recorrido completo, esta vez vamos a comparar 10 random walks

```python
all_walks = []
for i in range(10) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to NumPy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()

```

Esta grafica luce mal pues en el eje x se esta posicionando el valor de un random walk en particualar vs todos los valores que se obtubieron, en cambio debemos comparar 10 conjuntos de valores, en los que se grafica el numero de lanzamiento (index) vs el valor obtenido (el escalon) 

esto se logra facilmente con la traspuesta de la matriz

```python
# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

## incluir eventos/acciones que tienen una probabilidad de pasar en la simulacion : simular torpesa

supongamos que despues de lanzar el dado y avanzar, existe el peque√±o chance de 1% de caer y volver a 0, para esto se usa otra variable aleatoria, entre 0 y 1, si si valor es menor o igual a 0.01 (el 1% de 1) se devuelve al inicio

```python
all_walks = []
for i in range(25) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() <= 0.01 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
plt.clf()
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

aumenta el numero de simulaciones, y el numero de lanzamientos de dados, si el chance de caer aumenta la grfica parecera una gran entramado de ramas multicolor

Supongamos que aunqueremos saber la distribucion de el desenlace de estos experimentos, pero tenemos una estructura que contiene una lista de listas, una lista de random walks. vamos a extrer solo la ultima fila y a graficar un histograma

```python
print(np_aw_t)
ends = np_aw_t[-1,:]#todas las columnas(lanzamientos) pero solo la ultima fila () 
plt.clf()
plt.hist(ends)
plt.show()
```

Hagamos eso pero para 500 simulaciones del random walk y esta vez el chance de caer es solo del 0.1% en cada movimiento

```python

all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() <= 0.001 : # 0.1% chances e caer
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

np_aw_t = np.transpose(np.array(all_walks))
ends = np_aw_t[-1,:]
plt.clf()
plt.hist(ends)
plt.show()
```

Cual es la probabilidad de llegar mas alla del paso 60?

para calcular esto, en el arreglo de ends, donde estan todos los valores finales de las simulaciones, se cuentan cuantos valores estan en 60 o por encima a este valor se le divide el numero total de simulaciones

```python
count60 = len([i for i in ends if i >= 60])
numero_simulaciones = len(ends)

print(count60, 'entre', numero_simulaciones)
print('chances de ganar: ', (count60/numero_simulaciones)*100, '%')
```
