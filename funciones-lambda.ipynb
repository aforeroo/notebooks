# Funciones lambda

funciones sin nombre, tienen la forma

`lambda <argumentos>:<codigo>`

solo admiten una linea de codigo

```python
isPalindrome = lambda string: string == string[::-1]

print(isPalindrome('jaimito'))
```

# Funciones de orden superior

fucion que recibe como parametro a otra funcion

## Filter

se usa para filtrar o extraer datos de interes de una estructura iterable

se usa el siguiente formato 

`lista_filtrada = list( filter( lambda x:<expresion>, <iterable>) )`

por ejemplo, para extraer los valores impares de una lista de numeros

```python
lista = [1,2,3,7,9,11,14,15,18]
```

se puede hacer con comprehensions:

```python
solo_impares = [i for i in lista if i%2 !=0]
print(solo_impares)
```

o se puede hacer con filter:

```python
solo_impares = list (filter(lambda i: i%2 !=0, lista))
print(solo_impares)
```

## Map

para cada valor de un iterable: array, lista ... mapear(asociar) con un nuevo valor en otra lista

por ejemplo para una lista de numero enteros, mapear estos valores a otro arreglo donde se tengan el numero correspondiente elvado al cuadrado

```python
lista = [1,2,3,4,5]
```

Con comprehensions:

```python
lista_pow = [i**2 for i in lista]
print(lista_pow)
```

Con map:

```python
lista_pow = list(map(lambda i: i**2, lista))
print(lista_pow)
```

## Reduce

reduce todos los valores de una lista a un valor, aplicando una operacion a todos los valores

por ejemplo, si queremos concatenar todos los string en una lista de strings por separado dentro de un solo string

```python
lista= ['mi', 'nombre', 'es', 'Bond,', 'James', 'Bond']
```

con un for sencilo:

```python
final = ''
for word in lista: final += word + ' '
print(final)
```

con reduce:

```python
from functools import reduce
final = reduce(lambda word1, word2: word1 + ' ' + word2, lista )
print(final)
```

Algunas consultas de interes a los datos:

personas que dominand el lenguaje python

```python
import json# para mostrar claramente los diccionarios en consola

# nombre de los d
name_python_devs = [person for person in DATA if person['language']=='python']
for person in name_python_devs:
    print(json.dumps(person, indent=4))
```

Todas las personas que trabajan en platzi

```python
platzi_workers = [person for person in DATA if person['organization'] == 'Platzi']
for person in platzi_workers:
    print(json.dumps(person, indent=4))
```

Todas las personas mayores a 18 años

```python
mayores_edad = list(filter(lambda persona: persona['age']> 18, DATA))
for persona in mayores_edad:
    print(json.dumps(persona, indent=4))
```

Ahora solo quiero los nombres de los mayores de edad

```python
nombres_mayores = list(map(lambda persona : persona['name'], mayores_edad))
print(nombres_mayores)
```

Crear una nueva lista de diccionarios que contengan una nueva llave dependiendo si son mayores a 70 años o no

```python
new_DATA = list(map(lambda person: person | {'old':person['age'] > 70}, DATA ))
for person in new_DATA:
    print(json.dumps(person, indent=4))
```
