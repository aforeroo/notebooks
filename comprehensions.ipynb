# List comprehensions


```python
list_sqr = []
for i in range(1,101):
    if i%3 != 0:
        list_sqr.append(i**2)

print(list_sqr)
```

mismo resultado con list compregensions:

```python
squares = [i**2 for i in range(1,101) if i%3 != 0]
print(squares)
```

## Formula general

`list_name = [element for element in iterable if condition]`


`element` representa cada uno de los elemetos a agregar a la lista

`for element in iterabl` representa el ciclo de reoccorer otra estructura iterable de donde extraemos los elementos

`condition` condicion que debe cumplirse para que el elemento en la iteracion particular se agregue la lista

# Dictionary comprehensions

```python
dict = {}
for i in range(1,101):
    if i%3 != 0:
        dict[i] = i**3
print(dict)
```

Mismo resultado con dictionary comprehensions:

```python
dict = {i: i**3 for i in range(1,101) if i%3 != 0 }
print(dict)
```

## Formula general

`dict_name = [key: value(key) for key in iterable if condition]`


`key` representa la llave de cada uno de los elementos a agregar al diccionario

`value(key)` representa el valor (que puede depender de la llave  o no) asociada a la llave

`for key in iterabl` representa el ciclo de reoccorer otra estructura iterable de donde extraemos los valores de key

`condition` condicion que debe cumplirse para que la pareja llave:valor en la iteracion particular se agregue al diccionario
